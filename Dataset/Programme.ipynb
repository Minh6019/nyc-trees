```python
import numpy as np
import pandas as pd
import datetime
```

```python

url='data_100000.csv'
df_original=pd.read_csv(url, sep=',')
#df_original.head()

#df.info()
# for security->df
df = df_original.copy()
#df.head()

#drop columns 
df.drop(columns=['the_geom','state','boroname'], inplace=True)
df.head()

#total_nan_values = df.isnull().sum().sum()
#print ("total_nan_values=",total_nan_values) #34945

#total_na_values = df.isna().sum().sum()
#print ("total_na_values=",total_na_values) #34945

#df=df.dropna()
df = df.dropna(how='all')

#drop duplicated
df.drop_duplicates(subset ="tree_id", keep = False, inplace = True)

#replace ' ' by np.nan:
df = df.replace('', np.nan)
df = df.replace(' ', np.nan)

#convert date_time
df['created_at'] = pd.to_datetime(df['created_at'])
#df.head()
df.info()
```
