```python
def strip_strings(df):
    # strip strings #'the_geom', #'state'
    columns_to_strip = ['curb_loc',
                        'status',
                        'health',
                        'spc_latin',
                        'spc_common',
                        'steward',
                        'guards',
                        'sidewalk',
                        'user_type',
                        'problems',
                        'root_stone',
                        'root_grate',
                        'root_other',
                        'trnk_wire',
                        'trnk_light',
                        'trnk_other',
                        'brnch_ligh',
                        'brnch_shoe',
                        'brnch_othe',
                        'address',
                        'zip_city',
                        'boroname']
    
    df[columns_to_strip].apply(lambda x: x.str.strip())

def lower_strings(df):
    # strip strings
    columns_to_lower = ['curb_loc',
                        'status',
                        'health',
                        'spc_latin',
                        'spc_common',
                        'steward',
                        'guards',
                        'sidewalk',
                        'user_type',
                        'problems',
                        'root_stone',
                        'root_grate',
                        'root_other',
                        'trnk_wire',
                        'trnk_light',
                        'trnk_other',
                        'brnch_ligh',
                        'brnch_shoe',
                        'brnch_othe',
                        'address',
                        'zip_city',
                        'boroname']
    
    df[columns_to_lower].apply(lambda x: x.str.lower())
    
def bool_to_int(df):
    # strip strings
    columns_to_int = ['root_stone',
                      'root_grate',
                      'root_other',
                      'trnk_wire',
                      'trnk_light',
                      'trnk_other',
                      'brnch_ligh',
                      'brnch_shoe',
                      'brnch_othe']
                                                                          
    #df[columns_to_int].replace({True: 1, False: 0},inplace=True)
    df=df[columns_to_int]*1

```
